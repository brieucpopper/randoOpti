# Four Peaks

```python


import numpy as np
fitness = mlrose.FourPeaks(t_pct=0.3)



#print help on four peaksb

```

```python
#solve with random hill climbing
RUNS=50
graphLen=300

problem = mlrose.DiscreteOpt(length = 30, fitness_fn = fitness, maximize = True, max_val = 2)
actualFitnessCurves = []
actualFunctionCallsCurves=[]

bestFitnesses=[]

for averageOutRandomness in range(0,RUNS):

    best_state, best_fitness,fitness_curve = mlrose.random_hill_climb(problem, max_attempts=40, max_iters=graphLen, restarts=0, init_state=None, curve=True, random_state=averageOutRandomness)

    bestFitnesses.append(best_fitness)

    thisRunFitness=[x for x,y in fitness_curve]
    thisRunEvals=[y for x,y in fitness_curve]

    #make both arrays length graphLen by adding their last value to the end of the array enough times
    thisRunFitness = thisRunFitness + [thisRunFitness[-1]]*(graphLen-thisRunFitness.__len__())
    thisRunEvals = thisRunEvals + [thisRunEvals[-1]]*(graphLen-thisRunEvals.__len__())

    actualFitnessCurves.append(thisRunFitness)
    actualFunctionCallsCurves.append(thisRunEvals)


import matplotlib.pyplot as plt

print('the max best fitness is '+str(max(bestFitnesses)))

actualFinalFitnessCurve = np.zeros(actualFitnessCurves[0].__len__())
#fill it up by adding all the fitness curves then dividing by the number of fitness curves
for fitnessCurve in actualFitnessCurves:
    actualFinalFitnessCurve = actualFinalFitnessCurve + np.array(fitnessCurve)
actualFinalFitnessCurve = actualFinalFitnessCurve/actualFitnessCurves.__len__()

plt.plot(actualFinalFitnessCurve)




#legend the two lines on the plot, one is RHC fitness, the other is
plt.xlabel('Iterations')
plt.ylabel('Fitness')
# DO THIS plt.legend('RHC CLIMBING, 50 runs')
plt.title('RHC on Four Peaks(0.3 , 30)')
plt.legend(['RHC CLIMBING, 50 runs'])
plt.show()

```

```python
#solve with simulated annealing

RUNS=50
graphLen=800

problem = mlrose.DiscreteOpt(length = 30, fitness_fn = fitness, maximize = True, max_val = 2)
actualFitnessCurves = []
actualFunctionCallsCurves=[]

bestFitnesses=[]

for averageOutRandomness in range(0,RUNS):

    #TO MODIFY
    best_state, best_fitness,fitness_curve = mlrose.simulated_annealing(problem, schedule=mlrose.ExpDecay(), max_attempts=40, max_iters=graphLen, init_state=None, curve=True, random_state=averageOutRandomness)

    bestFitnesses.append(best_fitness)

    thisRunFitness=[x for x,y in fitness_curve]
    thisRunEvals=[y for x,y in fitness_curve]

    #make both arrays length graphLen by adding their last value to the end of the array enough times
    thisRunFitness = thisRunFitness + [thisRunFitness[-1]]*(graphLen-thisRunFitness.__len__())
    thisRunEvals = thisRunEvals + [thisRunEvals[-1]]*(graphLen-thisRunEvals.__len__())

    actualFitnessCurves.append(thisRunFitness)
    actualFunctionCallsCurves.append(thisRunEvals)



print('the max best fitness is '+str(max(bestFitnesses)))

actualFinalFitnessCurve = np.zeros(actualFitnessCurves[0].__len__())
#fill it up by adding all the fitness curves then dividing by the number of fitness curves
for fitnessCurve in actualFitnessCurves:
    actualFinalFitnessCurve = actualFinalFitnessCurve + np.array(fitnessCurve)
actualFinalFitnessCurve = actualFinalFitnessCurve/actualFitnessCurves.__len__()

plt.plot(actualFinalFitnessCurve)




#legend the two lines on the plot, one is RHC fitness, the other is
plt.xlabel('Iterations')
plt.ylabel('Fitness')

#TO MODIFY
plt.title('SA on Four Peaks(0.3 , 30)')
plt.legend(['Simulated A, 50 runs'])
plt.show()

```

# TSP

```python
#generate 100 cities with random x,y coords between 0 and 100
import random
random.seed(1)

num_cities=100
edges=[]
for i in range(num_cities):
    x=random.randint(0,100)
    y=random.randint(0,100)
    edges.append((x,y))
```

```python

import matplotlib.pyplot as plt



coords_list = edges

# Initialize fitness function object using coords_list
fitness_coords = mlrose.TravellingSales(coords = coords_list)

#visualize the graph by plotting in 2d space the vertices
import matplotlib.pyplot as plt

plt.scatter([x for x,y in coords_list], [y for x,y in coords_list])

plt.show()


problem_fit = mlrose.TSPOpt(length = len(edges), fitness_fn = fitness_coords, maximize=False)

a= mlrose.genetic_alg(problem_fit, random_state = 2)

print(a)

#Plot the solution by drawing lines between a[0][0], a[0][1] then a[0][1]], a[0][2], etc

for i in range(len(a[0])-1):
    plt.plot([coords_list[a[0][i]][0], coords_list[a[0][i+1]][0]], [coords_list[a[0][i]][1], coords_list[a[0][i+1]][1]])


# get random solution : sum of 0to1, 1to2, 2to3, 3to4, 4to5, 5to6, 6to7, 7to0
```

```python
# TO MODIFY
#solve with SA

RUNS=4
graphLen=4000

#TO MODIFY
problem = mlrose.TSPOpt(length = len(edges), fitness_fn = fitness_coords, maximize=False)


actualFitnessCurves = []
actualFunctionCallsCurves=[]

bestFitnesses=[]

for averageOutRandomness in range(0,RUNS):

    #TO MODIFY
    schedule=mlrose.ExpDecay(init_temp=2, exp_const=0.00003, min_temp=0.3)
    best_state, best_fitness,fitness_curve = mlrose.simulated_annealing(problem, schedule=schedule, max_attempts=200, max_iters=graphLen, init_state=None, curve=True, random_state=averageOutRandomness)

    bestFitnesses.append(best_fitness)

    thisRunFitness=[x for x,y in fitness_curve]
    thisRunEvals=[y for x,y in fitness_curve]

    #make both arrays length graphLen by adding their last value to the end of the array enough times
    thisRunFitness = thisRunFitness + [thisRunFitness[-1]]*(graphLen-thisRunFitness.__len__())
    thisRunEvals = thisRunEvals + [thisRunEvals[-1]]*(graphLen-thisRunEvals.__len__())

    actualFitnessCurves.append(thisRunFitness)
    actualFunctionCallsCurves.append(thisRunEvals)



print('the max best fitness is '+str(max(bestFitnesses)))

actualFinalFitnessCurve = np.zeros(actualFitnessCurves[0].__len__())
#fill it up by adding all the fitness curves then dividing by the number of fitness curves
for fitnessCurve in actualFitnessCurves:
    actualFinalFitnessCurve = actualFinalFitnessCurve + np.array(fitnessCurve)
actualFinalFitnessCurve = actualFinalFitnessCurve/actualFitnessCurves.__len__()

plt.plot(actualFinalFitnessCurve)




#legend the two lines on the plot, one is RHC fitness, the other is
plt.xlabel('Iterations')
plt.ylabel('Fitness')

#TO MODIFY
plt.title('RHCTODO (100,.15,40) on Four Peaks(0.3 , 30)')
plt.legend(['RHCTODO, 5 runs'])
plt.show()

```
